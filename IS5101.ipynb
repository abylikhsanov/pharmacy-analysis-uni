## Now let's investigate how the assortment level effects the sales

```python
df["Assortment"] = df["Assortment"].astype('category').cat.codes
```

```python
assort = df.groupby('Assortment')[["Sales", "Customers","CompetitionDistance"]].mean()
sales_customers_df = pd.DataFrame({'Assortment':assort.index,
                      'Sales':assort["Sales"], 'Customers': assort["Customers"], 'CompetitionDistance': assort['CompetitionDistance']}, 
                      columns=['Assortment', 'Sales', 'Customers', 'CompetitionDistance'])
store = pd.merge(sales_customers_df, df, on='Assortment')

assort.head()
```

```python
# Plotting the histogram
assort = df.groupby('DayOfWeek')[["Sales", "Customers"]].mean()
sales_customers_df = pd.DataFrame({'DayOfWeek':assort.index,
                      'Sales':assort["Sales"], 'Customers': assort["Customers"]}, 
                      columns=['Assortment', 'Sales', 'Customers'])
#store = pd.merge(sales_customers_df, df, on='DayOfWeek')

assort
```

```python
sizes = [assort['Sales'][1], assort['Sales'][5], assort['Sales'][6], assort['Sales'][4], assort['Sales'][0], assort['Sales'][2], assort['Sales'][3]]
labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red', 'green', 'blue']
#plt.pie(sizes, labels=labels, colors=colors, shadow=True, startangle=90)
patches, texts = plt.pie(sizes, colors=colors, startangle=90)
plt.legend(patches, labels, loc="best")
plt.legend(title="técnica")
plt.axis('equal')
plt.tight_layout()
```
